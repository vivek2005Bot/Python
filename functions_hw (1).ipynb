```python
#Given a list of numbers, return a new list with the square of each number
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda a:a**2,numbers))
print(squares)
```

```python
#Convert a list of temperatures from Celsius to Fahrenheit. The formula is f=c*(9/5)+32
celsius = [0, 10, 20, 30, 40]
fahrenheit = list(map(lambda c:c*(9/5)+32,celsius))
print(fahrenheit)
```

```python
#Given a list of strings, return a new list with the length of each string.
#example
#input->words = ["apple", "banana", "cherry"]
#output-> [5, 6, 6]
words = ["apple", "banana", "cherry"]
lengths = list(map(lambda w:len(w),words))
print(lengths)
```

```python
#Given two lists of numbers, return a new list containing the sum of corresponding elements.
list1 = [1, 2, 3]
list2 = [4, 5, 6]
sums = list(map(lambda a,b:a+b,list1,list2))
print(sums)
```
